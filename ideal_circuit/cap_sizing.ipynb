```python
# Specs
SNR_dB = 100.0             # target in-band SNR [dB]
f_s = 1e6                  # sampling frequency [Hz]
f_b = 1e3                  # signal bandwidth [Hz]
T = 300.0                  # temperature [K]

# Constants
k = 1.380649e-23           # Boltzmann [J/K]
Vamp = 1.35                # peak amplitude

# Compute C1
#OSR = f_s / (2*f_b)
vn_sq = 10**(-SNR_dB/10) * (Vamp**2) * 0.5
C1 = (k*T) / (512 * vn_sq)

# Compute 12C
C12 = C1 * 12

# Compute  4C
C4 = C1 * 4

# 12C = Cint1 and Cint2
# C4 = C2

# Sized Capacitors for the Circuit
print(f"C1 = {C1*1e15:.2f} fF   ({C1*1e12:.3f} pF)")
print(f"C12 = {C12*1e15:.2f} fF   ({C12*1e12:.3f} pF)")
print(f"C4 = {C4*1e15:.2f} fF   ({C4*1e12:.3f} pF)")
```
