```python
# PULSE Parameters
fs  = 1e6        # sampling frequency [Hz]
VDD = 3.0        # high level [V]
td  = 10e-9      # deadtime [s]
tr  = 1e-9       # rise time [s]
tf  = 1e-9       # fall time [s]

# Derived values
Ts = 1 / fs              # period [s]
PW = Ts/2 - td           # pulse width [s]

TD_phi1 = 0.0            # delay phi1
TD_phi2 = Ts / 2         # delay phi2

# Print SPICE PULSE sources
print(
    f"phi1 = PULSE(0 {VDD} "
    f"{TD_phi1*1e9:.0f}n {tr*1e9:.0f}n {tf*1e9:.0f}n "
    f"{PW*1e9:.0f}n {Ts*1e9:.0f}n)"
)

print(
    f"phi2 = PULSE(0 {VDD} "
    f"{TD_phi2*1e9:.0f}n {tr*1e9:.0f}n {tf*1e9:.0f}n "
    f"{PW*1e9:.0f}n {Ts*1e9:.0f}n)"
)
```
