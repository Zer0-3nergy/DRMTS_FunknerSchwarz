```python
from pygmid import Lookup as lk
import numpy as np
nmos = lk('./simulation/sg13_hv_nmos.mat')
pmos = lk('./simulation/sg13_hv_pmos.mat')
# parameters: VGS, VDS, VSB, L, W, NFING, ID, VT, GM, GMB, GDS, CGG, CGB, CGD, CGS, CDD, CSS, STH, SFL
# if not specified, minimum L, VDS=max(vgs)/2 and VSB=0 are used 
```

```python
# specs
id_spec = 8e-6
gm_id_spec = 12
L_spec = 0.13
```

```python
# gm calculation
gm = gm_id_spec * id_spec
print('gm =', round(gm/1e-3, 3), 'mS')
```

```python
vgs = pmos.look_upVGS(GM_ID=gm_id_spec, L=L_spec, VDS=1.65, VSB=0.0)
```

```python
# gm_gds look up and calculate gds from it
gm_gds = pmos.lookup('GM_GDS', GM_ID=gm_id_spec, L=L_spec, VDS=vgs, VSB=0)
gds = gm / gm_gds
print('gds =', round(gds/1e-6, 2), 'uS')
```

```python
# find the W of the mosfet
id_w = pmos.lookup('ID_W', GM_ID=gm_id_spec, L=L_spec, VDS=vgs, VSB=0)
w = id_spec / id_w
print('W =', round(w, 2), 'um, rounded W =', round(w*2)/2, 'um')
```
